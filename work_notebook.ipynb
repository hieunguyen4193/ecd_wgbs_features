```python
import pandas as pd 
import os

inputbed = "/media/hieunguyen/HNSD01/src/ecd_wgbs_features/hg19_CpG_clusters/CpG_clusters_whole_genome_radius_100_minCpG_5.bed"
df = pd.read_csv(inputbed, sep = "\t", header = None)
filename = inputbed.split("/")[-1]

for chr in range(1, 23):
    tmpdf = df[df[0] == chr].copy()
    tmpdf.to_csv(os.path.join(inputbed.replace(filename, ""), filename.replace(".bed", ""), f"chr{chr}.bed"))
```
